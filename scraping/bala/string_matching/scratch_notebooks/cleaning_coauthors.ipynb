```python
with open("../co_authors.json", "r") as file:
    data = json.load(file)
```

```python
corpus_researchers = data.keys()
```

```python
for key in data.keys():
    data[key] = list(filter(lambda x: x!=None , [r if r in corpus_researchers else None for r in set(data[key])]))
```

```python
for key in data.keys():
    if not data[key]:
        print(key)
```

```python
with open("../adjacency_list.json", "w") as file:
    json.dump(data, file)
```
