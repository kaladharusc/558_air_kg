```python
import networkx as nx
```

```python
G = nx.Graph()
```

```python
G.add_node('Abhinav Shrivastava')
```

```python
labels = {}
```

```python
labels["Abhinav Shrivastava"] = "Abhinav Shrivastava"
```

```python
for i, n in enumerate(set(data['Abhinav Shrivastava'])):
    print(i, n)
    G.add_edge('Abhinav Shrivastava', n)
    labels[n] = n
```

```python
import matplotlib.pyplot as plt
```

```python
pos=nx.spring_layout(G) # positions for all nodes
```

```python
len(labels) == len(G.nodes)
```

```python
G.nodes
```

```python
pos
```

```python
labels.items()
```

```python
fig = plt.figure()
fig.set_figheight(18)
fig.set_figwidth(18)
nx.draw_networkx_nodes(G,pos,
                       nodelist=list(G.nodes),
                       node_color='r',
                       node_size=5000,
                   alpha=0.8)
nx.draw_networkx_edges(G,pos,width=1.0,alpha=0.5)
nx.draw_networkx_labels(G,pos,labels,font_size=16)
plt.show()
```
